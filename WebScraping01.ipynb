```python
import urllib.request
URL='https://www.sunarp.gob.pe/mconsultas/RelacionS01.asp'
```

```python
Datos = urllib.request.urlopen(URL).read().decode()
Datos
```

### Trabajar en partes del HTML ###

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(Datos,'lxml')
tags = soup('input')
soup
```

```python
for tag in tags:
    print(tag.get('id'))
```

```python
tRazon = 'SOLUTIONS'
tRazon
```

```python
tRazon = soup.find("input", id="tRazon")
tRazon
```

```python
form = soup.find('form')
fields = form.findAll('input')
formdata = dict( (field.get('name'), field.get('value')) for field in fields)
formdata
```

```python
formdata['tRazon'] = u'SOLUTIONS'
```

```python
formdata
```

```python
##!pip install urlparse
##import urlparse

##!pip install urljoin
from urllib.parse import urljoin
```

```python
posturl = urljoin(URL, form['action'])
```

```python
posturl 
```

```python
import requests
s = requests.Session()
```

```python
r = s.post(posturl, data=formdata)
```
